```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, mean_squared_error, r2_score,
    roc_curve, auc, precision_recall_curve, average_precision_score,
    roc_auc_score
)

# Classification models
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC, LinearSVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import (
    RandomForestClassifier,
    ExtraTreesClassifier,
    GradientBoostingClassifier
)

from xgboost import XGBClassifier

import optuna
import shap
import lime
import lime.lime_tabular
import matplotlib.pyplot as plt
import seaborn as sns

# ============================================
# 1. Load and prepare data
# ============================================
data = pd.read_csv("/kaggle/input/crop-yield-prediction-dataset/yield_df.csv")

# Rename for convenience
data = data.rename(columns={
    "hg/ha_yield": "Yield",
    "average_rain_fall_mm_per_year": "Rainfall",
    "pesticides_tonnes": "Pesticides",
    "avg_temp": "Avg_Temp"
})

# Drop index-like column if present
if "Unnamed: 0" in data.columns:
    data = data.drop(columns=["Unnamed: 0"])

print("="*80)
print("DATA OVERVIEW")
print("="*80)
print("\nHead:")
display(data.head())
print("\nInfo:")
print(data.info())
print("\nDescribe:")
display(data.describe(include='all'))

# Encode categorical variables
le_country = LabelEncoder()
le_item = LabelEncoder()

data["Country_Encoded"] = le_country.fit_transform(data["Area"])
data["Item_Encoded"] = le_item.fit_transform(data["Item"])

# Features
feature_cols = ["Country_Encoded", "Item_Encoded", "Pesticides", "Avg_Temp", "Rainfall"]
X = data[feature_cols]

# Create CLASSIFICATION target (high vs low yield)
yield_threshold = data["Yield"].median()
print(f"\nYield threshold (median): {yield_threshold}")

data["Yield_High"] = (data["Yield"] >= yield_threshold).astype(int)
y_cls = data["Yield_High"]

print("\nClass distribution (0=low, 1=high):")
print(y_cls.value_counts(normalize=True))

# Correlation heatmap
plt.figure(figsize=(10, 8))
corr_matrix = data.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of Dataset')
plt.tight_layout()
plt.show()

# ============================================
# 2. Train/test split
# ============================================
X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(
    X, y_cls, test_size=0.2, random_state=42, stratify=y_cls
)

print(f"\nTraining set size: {len(X_train_cls)}")
print(f"Test set size: {len(X_test_cls)}")

# ============================================
# 3. Evaluation functions
# ============================================
def evaluate_classification(y_true, y_pred, y_pred_proba=None, model_name="Model"):
    """Complete evaluation with all metrics"""
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)
    
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    
    fnr = fn / (fn + tp) if (fn + tp) > 0 else 0.0
    fpr = fp / (fp + tn) if (fp + tn) > 0 else 0.0
    
    # AUC if probabilities available
    auc_roc = None
    if y_pred_proba is not None:
        try:
            auc_roc = roc_auc_score(y_true, y_pred_proba)
        except:
            auc_roc = None
    
    print(f"\n{'='*60}")
    print(f"{model_name}")
    print(f"{'='*60}")
    print(f"Accuracy : {acc:.4f}")
    print(f"Precision: {prec:.4f}")
    print(f"Recall   : {rec:.4f}")
    print(f"F1-score : {f1:.4f}")
    print(f"FNR      : {fnr:.4f}")
    print(f"FPR      : {fpr:.4f}")
    if auc_roc is not None:
        print(f"AUC-ROC  : {auc_roc:.4f}")
    print(f"\nConfusion Matrix:")
    print(f"  TN: {tn:5d}  FP: {fp:5d}")
    print(f"  FN: {fn:5d}  TP: {tp:5d}")
    
    return {
        "model": model_name,
        "accuracy": acc,
        "precision": prec,
        "recall": rec,
        "f1": f1,
        "fnr": fnr,
        "fpr": fpr,
        "auc_roc": auc_roc,
        "tn": tn,
        "fp": fp,
        "fn": fn,
        "tp": tp
    }

# ============================================
# 4. Define 10 classification models (BASE)
# ============================================
print("\n" + "="*80)
print("PHASE 1: BASELINE MODELS (BEFORE OPTUNA)")
print("="*80)

models_base = {
    "LogisticRegression": Pipeline([
        ("scaler", StandardScaler()),
        ("clf", LogisticRegression(max_iter=1000, random_state=42))
    ]),
    "KNN": Pipeline([
        ("scaler", StandardScaler()),
        ("clf", KNeighborsClassifier(n_neighbors=7))
    ]),
    "LinearSVC": Pipeline([
        ("scaler", StandardScaler()),
        ("clf", LinearSVC(max_iter=2000, random_state=42))
    ]),
    "SVC_RBF": Pipeline([
        ("scaler", StandardScaler()),
        ("clf", SVC(kernel="rbf", probability=True, random_state=42))
    ]),
    "GaussianNB": GaussianNB(),
    "DecisionTree": DecisionTreeClassifier(random_state=42),
    "RandomForest": RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1),
    "ExtraTrees": ExtraTreesClassifier(n_estimators=100, random_state=42, n_jobs=-1),
    "GradientBoosting": GradientBoostingClassifier(n_estimators=100, random_state=42),
    "XGBClassifier": XGBClassifier(
        n_estimators=100,
        learning_rate=0.1,
        max_depth=6,
        random_state=42,
        n_jobs=-1,
        eval_metric='logloss'
    )
}

# Train and evaluate all base models
results_base = []
trained_models_base = {}

for name, model in models_base.items():
    print(f"\nTraining {name}...")
    model.fit(X_train_cls, y_train_cls)
    y_pred = model.predict(X_test_cls)
    
    # Get probabilities if available
    y_pred_proba = None
    if hasattr(model, "predict_proba"):
        y_pred_proba = model.predict_proba(X_test_cls)[:, 1]
    elif hasattr(model, "decision_function"):
        y_pred_proba = model.decision_function(X_test_cls)
    
    metrics = evaluate_classification(y_test_cls, y_pred, y_pred_proba, f"{name} (Base)")
    results_base.append(metrics)
    trained_models_base[name] = {
        'model': model,
        'y_pred': y_pred,
        'y_pred_proba': y_pred_proba
    }

results_base_df = pd.DataFrame(results_base).set_index("model")
print("\n" + "="*80)
print("BASELINE MODELS SUMMARY (sorted by F1-score)")
print("="*80)
display(results_base_df.sort_values("f1", ascending=False))

# ============================================
# 5. OPTUNA OPTIMIZATION for all models
# ============================================
print("\n" + "="*80)
print("PHASE 2: OPTUNA HYPERPARAMETER TUNING")
print("="*80)

def objective_logreg(trial):
    C = trial.suggest_float("C", 1e-3, 10.0, log=True)
    model = Pipeline([
        ("scaler", StandardScaler()),
        ("clf", LogisticRegression(C=C, max_iter=1000, random_state=42))
    ])
    scores = cross_val_score(model, X_train_cls, y_train_cls, cv=3, scoring="f1", n_jobs=-1)
    return scores.mean()

def objective_knn(trial):
    n_neighbors = trial.suggest_int("n_neighbors", 3, 20)
    weights = trial.suggest_categorical("weights", ["uniform", "distance"])
    model = Pipeline([
        ("scaler", StandardScaler()),
        ("clf", KNeighborsClassifier(n_neighbors=n_neighbors, weights=weights))
    ])
    scores = cross_val_score(model, X_train_cls, y_train_cls, cv=3, scoring="f1", n_jobs=-1)
    return scores.mean()

def objective_svc(trial):
    C = trial.suggest_float("C", 1e-2, 10.0, log=True)
    gamma = trial.suggest_float("gamma", 1e-4, 1.0, log=True)
    model = Pipeline([
        ("scaler", StandardScaler()),
        ("clf", SVC(C=C, gamma=gamma, kernel="rbf", probability=True, random_state=42))
    ])
    scores = cross_val_score(model, X_train_cls, y_train_cls, cv=3, scoring="f1", n_jobs=-1)
    return scores.mean()

def objective_dt(trial):
    max_depth = trial.suggest_int("max_depth", 3, 20)
    min_samples_split = trial.suggest_int("min_samples_split", 2, 20)
    min_samples_leaf = trial.suggest_int("min_samples_leaf", 1, 10)
    model = DecisionTreeClassifier(
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        min_samples_leaf=min_samples_leaf,
        random_state=42
    )
    scores = cross_val_score(model, X_train_cls, y_train_cls, cv=3, scoring="f1", n_jobs=-1)
    return scores.mean()

def objective_rf(trial):
    n_estimators = trial.suggest_int("n_estimators", 100, 500)
    max_depth = trial.suggest_int("max_depth", 5, 30)
    min_samples_split = trial.suggest_int("min_samples_split", 2, 20)
    model = RandomForestClassifier(
        n_estimators=n_estimators,
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        random_state=42,
        n_jobs=-1
    )
    scores = cross_val_score(model, X_train_cls, y_train_cls, cv=3, scoring="f1", n_jobs=-1)
    return scores.mean()

def objective_et(trial):
    n_estimators = trial.suggest_int("n_estimators", 100, 500)
    max_depth = trial.suggest_int("max_depth", 5, 30)
    min_samples_split = trial.suggest_int("min_samples_split", 2, 20)
    model = ExtraTreesClassifier(
        n_estimators=n_estimators,
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        random_state=42,
        n_jobs=-1
    )
    scores = cross_val_score(model, X_train_cls, y_train_cls, cv=3, scoring="f1", n_jobs=-1)
    return scores.mean()

def objective_gb(trial):
    n_estimators = trial.suggest_int("n_estimators", 100, 500)
    learning_rate = trial.suggest_float("learning_rate", 0.01, 0.3, log=True)
    max_depth = trial.suggest_int("max_depth", 3, 10)
    model = GradientBoostingClassifier(
        n_estimators=n_estimators,
        learning_rate=learning_rate,
        max_depth=max_depth,
        random_state=42
    )
    scores = cross_val_score(model, X_train_cls, y_train_cls, cv=3, scoring="f1", n_jobs=-1)
    return scores.mean()

def objective_xgb(trial):
    params = {
        "n_estimators": trial.suggest_int("n_estimators", 100, 700),
        "max_depth": trial.suggest_int("max_depth", 3, 10),
        "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.3, log=True),
        "subsample": trial.suggest_float("subsample", 0.5, 1.0),
        "colsample_bytree": trial.suggest_float("colsample_bytree", 0.5, 1.0),
        "min_child_weight": trial.suggest_float("min_child_weight", 1.0, 10.0),
        "gamma": trial.suggest_float("gamma", 0.0, 5.0),
        "reg_lambda": trial.suggest_float("reg_lambda", 1e-3, 10.0, log=True),
        "reg_alpha": trial.suggest_float("reg_alpha", 1e-3, 10.0, log=True),
        "random_state": 42,
        "n_jobs": -1,
        "eval_metric": "logloss"
    }
    model = XGBClassifier(**params)
    scores = cross_val_score(model, X_train_cls, y_train_cls, cv=3, scoring="f1", n_jobs=-1)
    return scores.mean()

# Optimize each model
optuna.logging.set_verbosity(optuna.logging.WARNING)

optimized_models = {}
n_trials = 30

print("\n1. Optimizing LogisticRegression...")
study = optuna.create_study(direction="maximize")
study.optimize(objective_logreg, n_trials=n_trials, show_progress_bar=True)
best_params = study.best_trial.params
optimized_models["LogisticRegression"] = Pipeline([
    ("scaler", StandardScaler()),
    ("clf", LogisticRegression(C=best_params["C"], max_iter=1000, random_state=42))
])
print(f"Best F1 CV: {study.best_value:.4f}, Params: {best_params}")

print("\n2. Optimizing KNN...")
study = optuna.create_study(direction="maximize")
study.optimize(objective_knn, n_trials=n_trials, show_progress_bar=True)
best_params = study.best_trial.params
optimized_models["KNN"] = Pipeline([
    ("scaler", StandardScaler()),
    ("clf", KNeighborsClassifier(**best_params))
])
print(f"Best F1 CV: {study.best_value:.4f}, Params: {best_params}")

print("\n3. Optimizing SVC_RBF...")
study = optuna.create_study(direction="maximize")
study.optimize(objective_svc, n_trials=n_trials, show_progress_bar=True)
best_params = study.best_trial.params
optimized_models["SVC_RBF"] = Pipeline([
    ("scaler", StandardScaler()),
    ("clf", SVC(C=best_params["C"], gamma=best_params["gamma"], 
                kernel="rbf", probability=True, random_state=42))
])
print(f"Best F1 CV: {study.best_value:.4f}, Params: {best_params}")

print("\n4. Optimizing DecisionTree...")
study = optuna.create_study(direction="maximize")
study.optimize(objective_dt, n_trials=n_trials, show_progress_bar=True)
best_params = study.best_trial.params
optimized_models["DecisionTree"] = DecisionTreeClassifier(**best_params, random_state=42)
print(f"Best F1 CV: {study.best_value:.4f}, Params: {best_params}")

print("\n5. Optimizing RandomForest...")
study = optuna.create_study(direction="maximize")
study.optimize(objective_rf, n_trials=n_trials, show_progress_bar=True)
best_params = study.best_trial.params
optimized_models["RandomForest"] = RandomForestClassifier(**best_params, random_state=42, n_jobs=-1)
print(f"Best F1 CV: {study.best_value:.4f}, Params: {best_params}")

print("\n6. Optimizing ExtraTrees...")
study = optuna.create_study(direction="maximize")
study.optimize(objective_et, n_trials=n_trials, show_progress_bar=True)
best_params = study.best_trial.params
optimized_models["ExtraTrees"] = ExtraTreesClassifier(**best_params, random_state=42, n_jobs=-1)
print(f"Best F1 CV: {study.best_value:.4f}, Params: {best_params}")

print("\n7. Optimizing GradientBoosting...")
study = optuna.create_study(direction="maximize")
study.optimize(objective_gb, n_trials=n_trials, show_progress_bar=True)
best_params = study.best_trial.params
optimized_models["GradientBoosting"] = GradientBoostingClassifier(**best_params, random_state=42)
print(f"Best F1 CV: {study.best_value:.4f}, Params: {best_params}")

print("\n8. Optimizing XGBClassifier...")
study = optuna.create_study(direction="maximize")
study.optimize(objective_xgb, n_trials=50, show_progress_bar=True)
best_params = study.best_trial.params
best_params.update({"random_state": 42, "n_jobs": -1, "eval_metric": "logloss"})
optimized_models["XGBClassifier"] = XGBClassifier(**best_params)
print(f"Best F1 CV: {study.best_value:.4f}, Params: {best_params}")

# GaussianNB and LinearSVC don't have many hyperparameters to tune
optimized_models["GaussianNB"] = GaussianNB()
optimized_models["LinearSVC"] = Pipeline([
    ("scaler", StandardScaler()),
    ("clf", LinearSVC(max_iter=2000, random_state=42))
])

```

```python
# ============================================
# 6. Train and evaluate OPTIMIZED models
# ============================================
print("\n" + "="*80)
print("PHASE 3: OPTIMIZED MODELS (AFTER OPTUNA)")
print("="*80)

results_optuna = []
trained_models_optuna = {}

for name, model in optimized_models.items():
    print(f"\nTraining {name} (Optimized)...")
    model.fit(X_train_cls, y_train_cls)
    y_pred = model.predict(X_test_cls)
    
    y_pred_proba = None
    if hasattr(model, "predict_proba"):
        y_pred_proba = model.predict_proba(X_test_cls)[:, 1]
    elif hasattr(model, "decision_function"):
        y_pred_proba = model.decision_function(X_test_cls)
    
    metrics = evaluate_classification(y_test_cls, y_pred, y_pred_proba, f"{name} (Optuna)")
    results_optuna.append(metrics)
    trained_models_optuna[name] = {
        'model': model,
        'y_pred': y_pred,
        'y_pred_proba': y_pred_proba
    }

results_optuna_df = pd.DataFrame(results_optuna).set_index("model")
print("\n" + "="*80)
print("OPTIMIZED MODELS SUMMARY (sorted by F1-score)")
print("="*80)
display(results_optuna_df.sort_values("f1", ascending=False))

```

```python
# ============================================
# 7. COMPARISON: Before vs After Optuna
# ============================================
print("\n" + "="*80)
print("COMPARISON: BASE vs OPTUNA")
print("="*80)

comparison_data = []
for name in models_base.keys():
    base_metrics = results_base_df.loc[f"{name} (Base)"]
    optuna_metrics = results_optuna_df.loc[f"{name} (Optuna)"]
    
    comparison_data.append({
        "Model": name,
        "Base_F1": base_metrics["f1"],
        "Optuna_F1": optuna_metrics["f1"],
        "Improvement": optuna_metrics["f1"] - base_metrics["f1"],
        "Base_Accuracy": base_metrics["accuracy"],
        "Optuna_Accuracy": optuna_metrics["accuracy"]
    })

comparison_df = pd.DataFrame(comparison_data)
display(comparison_df.sort_values("Optuna_F1", ascending=False))

# Visualization
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# F1-score comparison
ax = axes[0]
models_names = comparison_df["Model"]
x = np.arange(len(models_names))
width = 0.35
ax.bar(x - width/2, comparison_df["Base_F1"], width, label='Base', alpha=0.8)
ax.bar(x + width/2, comparison_df["Optuna_F1"], width, label='Optuna', alpha=0.8)
ax.set_xlabel('Models')
ax.set_ylabel('F1-Score')
ax.set_title('F1-Score: Base vs Optuna')
ax.set_xticks(x)
ax.set_xticklabels(models_names, rotation=45, ha='right')
ax.legend()
ax.grid(True, alpha=0.3)

# Improvement chart
ax = axes[1]
colors = ['green' if x > 0 else 'red' for x in comparison_df["Improvement"]]
ax.barh(models_names, comparison_df["Improvement"], color=colors, alpha=0.7)
ax.set_xlabel('F1-Score Improvement')
ax.set_title('Improvement After Optuna Tuning')
ax.axvline(x=0, color='black', linestyle='--', linewidth=0.8)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

```

```python
# ============================================
# 8. ROC AUC & Precision-Recall Curves
# ============================================
print("\n" + "="*80)
print("PHASE 4: ROC AUC & PRECISION-RECALL CURVES")
print("="*80)

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# ROC Curves - Base models
ax = axes[0, 0]
for name, model_data in trained_models_base.items():
    if model_data['y_pred_proba'] is not None:
        fpr, tpr, _ = roc_curve(y_test_cls, model_data['y_pred_proba'])
        roc_auc = auc(fpr, tpr)
        ax.plot(fpr, tpr, label=f'{name} (AUC={roc_auc:.3f})', alpha=0.7)
ax.plot([0, 1], [0, 1], 'k--', label='Random')
ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate')
ax.set_title('ROC Curves - Base Models')
ax.legend(loc='lower right', fontsize=8)
ax.grid(True, alpha=0.3)

# ROC Curves - Optuna models
ax = axes[0, 1]
for name, model_data in trained_models_optuna.items():
    if model_data['y_pred_proba'] is not None:
        fpr, tpr, _ = roc_curve(y_test_cls, model_data['y_pred_proba'])
        roc_auc = auc(fpr, tpr)
        ax.plot(fpr, tpr, label=f'{name} (AUC={roc_auc:.3f})', alpha=0.7)
ax.plot([0, 1], [0, 1], 'k--', label='Random')
ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate')
ax.set_title('ROC Curves - Optuna Models')
ax.legend(loc='lower right', fontsize=8)
ax.grid(True, alpha=0.3)

# Precision-Recall Curves - Base models
ax = axes[1, 0]
for name, model_data in trained_models_base.items():
    if model_data['y_pred_proba'] is not None:
        precision, recall, _ = precision_recall_curve(y_test_cls, model_data['y_pred_proba'])
        avg_prec = average_precision_score(y_test_cls, model_data['y_pred_proba'])
        ax.plot(recall, precision, label=f'{name} (AP={avg_prec:.3f})', alpha=0.7)
ax.set_xlabel('Recall')
ax.set_ylabel('Precision')
ax.set_title('Precision-Recall Curves - Base Models')
ax.legend(loc='lower left', fontsize=8)
ax.grid(True, alpha=0.3)

# Precision-Recall Curves - Optuna models
ax = axes[1, 1]
for name, model_data in trained_models_optuna.items():
    if model_data['y_pred_proba'] is not None:
        precision, recall, _ = precision_recall_curve(y_test_cls, model_data['y_pred_proba'])
        avg_prec = average_precision_score(y_test_cls, model_data['y_pred_proba'])
        ax.plot(recall, precision, label=f'{name} (AP={avg_prec:.3f})', alpha=0.7)
ax.set_xlabel('Recall')
ax.set_ylabel('Precision')
ax.set_title('Precision-Recall Curves - Optuna Models')
ax.legend(loc='lower left', fontsize=8)
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# ============================================
# 9. XAI - SHAP for ALL OPTUNA MODELS
# ============================================
print("\n" + "="*80)
print("PHASE 5: XAI - SHAP ANALYSIS FOR ALL MODELS")
print("="*80)

# Sample for SHAP computation
X_sample = X_train_cls.sample(n=min(500, len(X_train_cls)), random_state=42)
X_sample_test = X_test_cls.sample(n=min(100, len(X_test_cls)), random_state=42)

tree_based_models = ["DecisionTree", "RandomForest", "ExtraTrees", 
                     "GradientBoosting", "XGBClassifier"]

for model_name in tree_based_models:
    print(f"\n{'='*60}")
    print(f"SHAP Analysis: {model_name} (Optuna)")
    print(f"{'='*60}")
    
    model = trained_models_optuna[model_name]['model']
    
    try:
        # Create explainer
        explainer = shap.TreeExplainer(model)
        shap_values = explainer.shap_values(X_sample)
        
        # Handle different SHAP value formats
        if isinstance(shap_values, list):
            shap_values_class1 = shap_values[1]
        else:
            shap_values_class1 = shap_values
        
        # 1. Beeswarm Summary Plot
        print(f"\n{model_name}: Beeswarm Summary Plot")
        plt.figure(figsize=(10, 6))
        shap.summary_plot(shap_values_class1, X_sample, 
                         feature_names=feature_cols, show=False)
        plt.title(f'{model_name} - SHAP Beeswarm Summary', fontsize=14, fontweight='bold')
        plt.tight_layout()
        plt.show()
        plt.close()
        
        # 2. Bar plot showing mean absolute SHAP values
        print(f"\n{model_name}: Feature Importance (Mean |SHAP|)")
        plt.figure(figsize=(10, 6))
        shap.summary_plot(shap_values_class1, X_sample, 
                         feature_names=feature_cols, 
                         plot_type="bar", show=False)
        plt.title(f'{model_name} - Feature Importance (Mean Absolute SHAP)', 
                 fontsize=14, fontweight='bold')
        plt.tight_layout()
        plt.show()
        plt.close()
        
        # 3. Waterfall plots for multiple test instances (INDIVIDUAL PLOTS)
        print(f"\n{model_name}: Waterfall Plots (Multiple Test Instances)")
        shap_values_test = explainer.shap_values(X_sample_test)
        
        if isinstance(shap_values_test, list):
            shap_values_test_class1 = shap_values_test[1]
            expected_value = explainer.expected_value[1]
        else:
            shap_values_test_class1 = shap_values_test
            expected_value = explainer.expected_value
        
        # Create INDIVIDUAL waterfall plots (one per figure)
        n_waterfall_instances = min(5, len(X_sample_test))
        
        for idx in range(n_waterfall_instances):
            print(f"  Generating waterfall plot for instance {idx}...")
            
            # Create new figure for each waterfall
            fig, ax = plt.subplots(figsize=(12, 6))
            
            # Create explanation object for waterfall plot
            explanation = shap.Explanation(
                values=shap_values_test_class1[idx],
                base_values=expected_value,
                data=X_sample_test.iloc[idx].values,
                feature_names=feature_cols
            )
            
            # Plot waterfall using new API
            shap.plots.waterfall(explanation, max_display=10, show=False)
            plt.title(f'{model_name} - Waterfall Plot (Test Instance {idx})\n' + 
                     f'Actual: {y_test_cls.iloc[X_sample_test.index[idx]]}, ' +
                     f'Predicted: {trained_models_optuna[model_name]["y_pred"][X_sample_test.index[idx]]}',
                     fontsize=12, fontweight='bold')
            plt.tight_layout()
            plt.show()
            plt.close()
        
        # 4. Force Plot visualization (matplotlib for display)
        print(f"\n{model_name}: Force Plots (First 3 instances)")
        for idx in range(min(3, len(X_sample_test))):
            try:
                plt.figure(figsize=(14, 3))
                shap.force_plot(
                    expected_value,
                    shap_values_test_class1[idx],
                    X_sample_test.iloc[idx],
                    feature_names=feature_cols,
                    matplotlib=True,
                    show=False
                )
                plt.title(f'{model_name} - Force Plot (Test Instance {idx})', 
                         fontsize=12, fontweight='bold')
                plt.tight_layout()
                plt.show()
                plt.close()
            except Exception as e:
                print(f"  Could not create matplotlib force plot for instance {idx}: {e}")
        
        # 5. Stacked Force Plot (Interactive HTML)
        print(f"\n{model_name}: Stacked Force Plot (Interactive - Saved as HTML)")
        try:
            shap.initjs()
            force_plot_stacked = shap.force_plot(
                expected_value,
                shap_values_test_class1[:30],
                X_sample_test.iloc[:30],
                feature_names=feature_cols
            )
            
            # Save as HTML
            html_filename = f'force_plot_stacked_{model_name}.html'
            shap.save_html(html_filename, force_plot_stacked)
            print(f"   Stacked force plot saved as '{html_filename}'")
            print(f"  Open this file in a browser to see interactive stacked visualization")
        except Exception as e:
            print(f"  Could not create stacked force plot: {e}")
        
        # 6. Decision Plot (shows prediction paths)
        print(f"\n{model_name}: Decision Plot")
        try:
            plt.figure(figsize=(12, 8))
            shap.decision_plot(
                expected_value,
                shap_values_test_class1[:20],
                X_sample_test.iloc[:20],
                feature_names=feature_cols,
                show=False
            )
            plt.title(f'{model_name} - Decision Plot (20 instances)', 
                     fontsize=14, fontweight='bold')
            plt.tight_layout()
            plt.show()
            plt.close()
        except Exception as e:
            print(f"  Could not create decision plot: {e}")
        
        # 7. Dependence plots for top 2 features
        print(f"\n{model_name}: Dependence Plots (Top 2 Features)")
        try:
            # Get feature importance
            feature_importance = np.abs(shap_values_class1).mean(axis=0)
            top_features_idx = np.argsort(feature_importance)[-2:]
            
            fig, axes = plt.subplots(1, 2, figsize=(16, 5))
            
            for i, feat_idx in enumerate(top_features_idx):
                plt.sca(axes[i])
                shap.dependence_plot(
                    feat_idx,
                    shap_values_class1,
                    X_sample,
                    feature_names=feature_cols,
                    show=False
                )
                axes[i].set_title(f'Dependence: {feature_cols[feat_idx]}', 
                                 fontsize=12, fontweight='bold')
            
            plt.suptitle(f'{model_name} - SHAP Dependence Plots', 
                        fontsize=14, fontweight='bold', y=1.02)
            plt.tight_layout()
            plt.show()
            plt.close()
        except Exception as e:
            print(f"  Could not create dependence plots: {e}")
        
        print(f"\n Completed SHAP analysis for {model_name}")
        print("-" * 60)
        
    except Exception as e:
        print(f" Error computing SHAP for {model_name}: {str(e)}")
        import traceback
        traceback.print_exc()

print("\n" + "="*80)
print("SHAP ANALYSIS COMPLETED FOR ALL MODELS")
print("="*80)
print("\nSummary of XAI visualizations created for each model:")
print("  1. Beeswarm Summary Plot - Feature impact distribution")
print("  2. Bar Plot - Mean absolute SHAP values")
print("  3. Waterfall Plots (5 instances) - Individual prediction breakdowns")
print("  4. Force Plots (3 instances) - Push/pull visualization")
print("  5. Stacked Force Plot (HTML) - Interactive multi-instance view")
print("  6. Decision Plot - Cumulative feature effects")
print("  7. Dependence Plots - Feature interaction effects")
print("="*80)
```

```python
# ============================================
# 10. Find BEST MODEL and apply LIME
# ============================================
print("\n" + "="*80)
print("PHASE 6: BEST MODEL SELECTION & LIME ANALYSIS")
print("="*80)

best_model_name = results_optuna_df["f1"].idxmax()
best_model_name_clean = best_model_name.replace(" (Optuna)", "")
best_model = trained_models_optuna[best_model_name_clean]['model']

print(f"\nBest Model: {best_model_name}")
print(f"Best F1-Score: {results_optuna_df.loc[best_model_name, 'f1']:.4f}")
print(f"\nFull metrics:")
display(results_optuna_df.loc[best_model_name])

# LIME Explainer
print("\n" + "="*80)
print(f"LIME EXPLANATION FOR {best_model_name}")
print("="*80)

# Create LIME explainer
lime_explainer = lime.lime_tabular.LimeTabularExplainer(
    X_train_cls.values,
    feature_names=feature_cols,
    class_names=['Low Yield', 'High Yield'],
    mode='classification',
    random_state=42
)

# Explain several test instances
n_lime_instances = 5
print(f"\nGenerating LIME explanations for {n_lime_instances} test instances...")

for i in range(n_lime_instances):
    print(f"\n--- LIME Explanation for Test Instance {i} ---")
    
    # Get prediction function
    if hasattr(best_model, "predict_proba"):
        predict_fn = best_model.predict_proba
    else:
        # For models without predict_proba, wrap decision_function
        def predict_fn(X):
            scores = best_model.decision_function(X)
            # Convert to probabilities
            exp_scores = np.exp(scores - np.max(scores))
            probs = exp_scores / exp_scores.sum()
            return np.column_stack([1 - probs, probs])
    
    # Generate explanation
    exp = lime_explainer.explain_instance(
        X_test_cls.iloc[i].values,
        predict_fn,
        num_features=len(feature_cols)
    )
    
    # Show in notebook
    print(f"\nActual class: {y_test_cls.iloc[i]}")
    print(f"Predicted class: {trained_models_optuna[best_model_name_clean]['y_pred'][i]}")
    
    # Show explanation
    fig = exp.as_pyplot_figure()
    plt.title(f'LIME Explanation - Test Instance {i}')
    plt.tight_layout()
    plt.show()
    
    # Print text explanation
    print("\nFeature contributions:")
    for feature, weight in exp.as_list():
        print(f"  {feature}: {weight:.4f}")

```
